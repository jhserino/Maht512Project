```python
movies2 = pd.read_csv("tmdbData.csv",index_col=0)
movies2 = movies2.drop(columns=['id','imdb_id','homepage','overview','tagline','keywords','original_title'])
movies2.head(2)
```

We split up actors into at much three of each

```python
movies2['LeadActor']=movies2['cast'].str.split('|').str[0]
movies2['SecondActor']=movies2['cast'].str.split('|').str[1]
movies2['ThirdActor']=movies2['cast'].str.split('|').str[2]
```

```python
movies2.head(2)
```

We split up genres and production companies into at much three of each

```python
movies2['PrimaryGenre']=movies2['genres'].str.split('|').str[0]
movies2['SecondaryGenre']=movies2['genres'].str.split('|').str[1]
movies2['ThirdGenre']=movies2['genres'].str.split('|').str[2]
movies2['PrimaryProd']=movies2['production_companies'].str.split('|').str[0]
movies2['SecondaryProd']=movies2['production_companies'].str.split('|').str[1]
movies2['ThirdProd']=movies2['production_companies'].str.split('|').str[2]
```

```python
movies2.head(2)
```

We split up the string for release date into month and day turning them into integers, not factoring in release year

```python
movies2['ReleaseMonth']=movies2['release_date'].str.split('-').str[1]
movies2['ReleaseDay']=movies2['release_date'].str.split('-').str[2]
movies2['ReleaseMonth'] = [int(string) for string in movies2['ReleaseMonth']]
movies2['ReleaseDay'] = [int(string) for string in movies2['ReleaseDay']]
```

```python
movies2.head(2)
```

We add the release month to the release day/30 to try to get numerical approximations for release dates

```python
movies2['CombinedRelease'] = [x + y/30 for x, y in zip(movies2['ReleaseMonth'],movies2['ReleaseDay'])]
```

```python
movies2.head(2)
```

We see that the only features with missing values are secondary genre, third genre, second prod and third prod

```python
movies2.head(2)
```

```python
#Function that turns categorical columns into numerical columns
def dummyCode(feature_column):
    strings = {}
    dummyValue = 0

    # Convert strings to integers using a loop
    for string in feature_column:
        if string not in strings:
            strings[string] = dummyValue
            dummyValue += 1
        
    feature_column = [strings[string] for string in feature_column]
    return feature_column  
```

```python
movies2['PrimaryGenre'] = dummyCode(movies2['PrimaryGenre'])
movies2['SecondaryGenre'] = dummyCode(movies2['SecondaryGenre'])
movies2['ThirdGenre'] = dummyCode(movies2['ThirdGenre'])

movies2['PrimaryProd'] = dummyCode(movies2['PrimaryProd'])
movies2['SecondaryProd'] = dummyCode(movies2['SecondaryProd'])
movies2['ThirdProd'] = dummyCode(movies2['ThirdProd'])

movies2['director'] = dummyCode(movies2['director'])

movies2['LeadActor'] = dummyCode(movies2['LeadActor'])
movies2['SecondActor'] = dummyCode(movies2['SecondActor'])
movies2['ThirdActor'] = dummyCode(movies2['ThirdActor'])

movies2['popularity_level'] = dummyCode(movies2['popularity_level'])
```

```python
# Creating feature frame out of all quantitative values, and dropping 'Superbowl' column
X = movies2.drop(['revenue','profit'],axis = 1)
# Creating classification
y = movies2['revenue']

# Importing modules from sci-kit for working with predicting data
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

# Creating a test set with 25% of the data
X_train, X_test, y_train, y_test = train_test_split(X, y,
  test_size=0.25,
  shuffle=True,
  random_state=3
)

# Training knn classifier to compute test accuracy
knn = KNeighborsClassifier(n_neighbors=8)
knn.fit(X_train, y_train)    # fit only to train set
acc = knn.score(X_test, y_test)    # score only on test set
print(f"test accuracy is {acc:.1%}")

# Training new variables in order to compute predictor accuracy
n,d = X.shape
# Get vector of predictions for the training set:
yhat = knn.predict(X)    
acc = sum(yhat == y) / n    # fraction of correct predictions
print(f"accuracy is {acc:.1%}")
```

```python
from sklearn.linear_model import Lasso
from sklearn.linear_model import LinearRegression

lass = Lasso(alpha=0.05)
lass.fit(X_train, y_train)

pd.DataFrame( {
    "feature": X.columns,
    "LASSO": lass.coef_
    } )
```

```python
# Get the locations (indices) of the very small weights:
zeroed = np.nonzero( np.abs(lass.coef_) < 1e-4 )
# Names of the corresponding columns:
dropped = X.columns[zeroed].values

X_train_reduced = X_train.drop(dropped, axis=1)
X_test_reduced = X_test.drop(dropped, axis=1)

X_train_reduced.head()
```

```python
lm = LinearRegression()
lm.fit(X_train, y_train)

print(f"original linear model score: {lm.score(X_test, y_test):.4f}")

lm.fit(X_train_reduced, y_train)
Rsq = lm.score(X_test_reduced, y_test)


print(f"reduced linear model score: {Rsq:.4f}")
```

```python
#Function that turns categorical columns into numerical columns
def removeCommas(feature_column):
    for string in feature_column:
        feature_column[string] = feature_column.str.replace(',','')
    return feature_column  

def removeDollarSigns(feature_column):
    for string in feature_column:
        feature_column[string] = feature_column.str.replace('$','').astpye(int)
    return 
```

```python
movies1['Domestic Box Office'] = removeCommas(movies1['Domestic Box Office'])
movies1['Domestic Box Office'] = removeDollarSigns(movies1['Domestic Box Office'])
```
